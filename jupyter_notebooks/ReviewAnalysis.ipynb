```python
from feature_engine.encoding import OneHotEncoder
encoder = OneHotEncoder(variables=df.columns[df.dtypes=='object'].to_list(), drop_last=False)
df_ohe = encoder.fit_transform(df)
print(df_ohe.shape)
df_ohe.head(3)
```

- Transforming categorical variables in numerical to perform correlation study
- Choosing the best method after studying the variables
- Methods that were thought by the developer:
  1. Label Encoding = Label encoding assigns a unique numerical label to each category in a categorical variable. This method is suitable for ordinal variables where there is an inherent order among the categories.
  2. One-Hot Encoding = One-hot encoding creates binary columns for each category, representing the presence or absence of that category. It is suitable for nominal categorical variables.
  3. Frequency Encoding = Frequency encoding replaces each category with the frequency (count) of that category in the dataset. This can be useful when the frequency of categories is relevant information.
- Variables does not follow a specific order, frequency is not relevant information for the categorical variables. Hence method chosen by the developer was OneHotEncoding()

```python
corr_spearman = df_ohe.corr(method='spearman')['review_overall'].sort_values(key=abs, ascending=False).head(10)
corr_spearman
```
